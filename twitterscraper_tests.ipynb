```python
import snscrape.modules.twitter as sntwitter
import csv
import pandas as pd
from nltk.tokenize import TweetTokenizer
import nltk
from nltk.corpus import stopwords

tokenizer = TweetTokenizer()
```

```python
def scrape_users(users, zeitraum):
    
    tweets_list = []

    for user in users:
        
        conditions = 'from:{} {}'.format(user, zeitraum)

        for i,tweet in enumerate(sntwitter.TwitterSearchScraper(conditions).get_items()): #declare a username
            tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username]) #declare the attributes to be returned
        
    return pd.DataFrame(tweets_list, columns=['date', 'id', 'text', 'username'])

        # tweets_df["tokenized_text"] = tokenizer.tokenize(tweets_df["text"])

        # add to csv
        # tweets_df.to_csv('csvtesttt.csv', mode='a', header=False)
```

```python
users = ["OlafScholz", "jensspahn", "c_lindner", "SWagenknecht"]
zeitraum = "since:2017-10-24 until:2017-10-27" # 2021

df = scrape_users(users, zeitraum)
df['text_tokenized'] = df.apply(lambda x: tokenizer.tokenize(x['text']),axis=1)
```

```python
# "tweet_tokenized" aufräumen: lowercase, stopwords

df['text_tokenized'] =  df['text_tokenized'].apply(lambda x: list(map(str.lower, x)))

# w.lower() for w in tokens
stopwords = stopwords.words('german')

df['text_without_stopwords'] = df['text_tokenized'].apply(lambda words: [word for word in words if word not in stopwords])
```

```python
import spacy
nlp = spacy.load("de_core_news_sm")

df['text_without_stopwords_spacy'] = df['text_without_stopwords'].apply(lambda words: [nlp(word) for word in words])
```
